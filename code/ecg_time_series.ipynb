```python
def cosine_similarity(x, y):
    """
    두 벡터 x, y의 코사인 유사도를 계산하여 반환.
    x, y: 1차원 numpy 배열
    """
    denom = (np.linalg.norm(x) * np.linalg.norm(y))
    if denom == 0:
        return 0.0
    return np.dot(x, y) / denom
```

```python
num_samples = 100

cos_sim_matrix = np.zeros((num_samples, num_samples))

for i in range(num_samples):
    for j in range(num_samples):
        cos_sim_matrix[i, j] = cosine_similarity(ecg_data[i], ecg_data[j])
```

```python
cos_sim_matrix.shape
```

```python
groups = [2,2,6,2,2,2,3,1,6,2,2,2,6,2,2,1,1,6,2,6,1,6,6,1,6,1,2,2,2,1,2,6,4,5,3,1,2,2,2,6,1,5,2,3,1,5,5,5,3,6,5,3,1,5,5,1,4,4,6,1,2,2,1,2,6,1,2,1,2,6,5,4,6,4,2,1,1,1,1,6,1,2,6,2,1,5,4,2,5,6,2,4,2,2,1,1,2,6,2,4]
unique_groups = np.unique(groups)  # ['A', 'B', 'C'] 등이 나온다고 가정

# 결과를 보기 좋게 저장하기 위한 딕셔너리
# (group1, group2) -> 평균 코사인 유사도
avg_cosine_dict = {}

for g1 in unique_groups:
    for g2 in unique_groups:
        # g1, g2 집단의 index 추출
        idx_g1 = np.where(groups == g1)[0]
        idx_g2 = np.where(groups == g2)[0]
        
        # 두 집단 각각에서 모든 쌍(i, j)에 대해 코사인 유사도 평균
        # (g1 == g2 인 경우, 집단 내 평균 유사도)
        pair_sim = []
        for i in idx_g1:
            for j in idx_g2:
                if(i!=j):
                # i != j 로 해서 자기 자신 제외할지 여부는 상황에 맞게 선택
                # 여기서는 동일 관측치(i==j)도 포함
                    pair_sim.append(cos_sim_matrix[i, j])
                        
        mean_sim = np.mean(pair_sim) if len(pair_sim) > 0 else 0.0
        avg_cosine_dict[(g1, g2)] = mean_sim
        
print("=== 평균 코사인 유사도 표 (집단별) ===")
for g1 in unique_groups:
    for g2 in unique_groups:
        print(f"Group {g1} vs Group {g2} : {avg_cosine_dict[(g1, g2)]:.4f}")
    print("")

# -----------------------------
# 6) 예시: 특정 집단 쌍의 유사도가 1에 가깝다면?
# -----------------------------
# 예: A vs B 집단의 평균 유사도가 1에 가깝다면, 두 집단 간
#    '방향 패턴이 매우 비슷하다'고 볼 수 있음.
print("=== 집단 쌍 간 평균 유사도 ===")
for (g1, g2), val in avg_cosine_dict.items():
    print(f"{g1} - {g2} : {val:.4f}")
```
